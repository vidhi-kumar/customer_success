### Lets apply some transformations and feature engineering over the dataframe

#### "price: Unit price. Numeric. Product price per unit in Turkish Liras (TL)."
#### We will convert TL to American Dollars for our convinience: 1 Turkish Liras: 0.029 United States Dollar

```python
df['price'] = df['price'] * 0.029
```

#### Lets calculate total price using quantity and price which will help us calculate the revenue

```python
df['total_price'] = df['quantity'] * df['price']
```

#### We will convert the invoice_date field to date time datatype to have better grip over it

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format="%d/%m/%Y")
```

#### Lets decompose the date field into month, year, day, quarter

```python
df['invoice_year'] = df['invoice_date'].dt.year
df['invoice_day'] = df['invoice_date'].dt.day
df['invoice_month'] = df['invoice_date'].dt.month_name()
df['invoice_weekday'] = df['invoice_date'].dt.weekday #0 -> monday, 6 -> sunday
df['invoice_quarter'] = df['invoice_date'].dt.quarter
df['invoice_year_month'] = df['invoice_date'].dt.to_period('M')
df['invoice_year_quarter'] = df['invoice_date'].dt.to_period('Q')
```

### Time for visualisations

#### Let check the correlation matrix to see if we have any strong correlation 

```python
# correlation works with numerical values, so we will use only those columns for the correlation matrix
df_numeric = df[['age', 'quantity', 'price', 'total_price']]
correlation_matrix = df_numeric.corr()
# print(correlation_matrix)
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.show()
```

#### Nothing much here, since total_price is calculated via price and quantity, we see strong correlation

#### Lets check the distribution of values across the numeric values

```python
df_numeric.hist(figsize=(10, 6), bins=25)
plt.show()
```

#### We can observe the following:
1. Age is distributed evenly, we have teenagers as well as people approaching 70 in similar numbers
2. Quantity of products purchased is also even speard over 1 to 5
3. Price of majority of products are below the 1000 dollars mark. We have a very small fraction of products having price around 5.5K Dollars
4. Total price varies with the price and quantity. we can see some purchases from customer for over 25k Dollars
